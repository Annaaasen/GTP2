# Oppgave 1

Vi startet med å installere og importerer ```aitextgen``` biblioteket, som inneholder OpenAI sin GTP-2 modell. 

````python
!pip install -q aitextgen
from aitextgen import aitextgen
````

For å benytte seg av den standardtrente GTP-2 modellen, satt vi opp en instanse av ```aitextgen```. Underveis er det verdt å benytte seg av eventuelle akkseleratorer (CPU eller lignende) for å trene og generere tekst med modellen, da dette er svært ressurskrevende. 

````python
ai = aitextgen(to_gpu=True)
````

Vi ønsket nå å undersøke hvor god modellen var, derfor brukte vi ```generate```-funksjonen til å generere en tekst hver (antall teksten den skriver ut kan justeres med _n_ parameteren) med følgende ulike _prompts_:
* "Trump is president because"
* "Climate change is"
* "The coronavirus is"
* "What is Black Lives Matter?"
* "What is white supremacy?"

````python
ai.generate(n=1, prompt='Trump is president because', max_length=100, seed=10)
print('-----')
ai.generate(n=1, prompt='Climate change is', max_length=100, seed=11)
print('-----')
ai.generate(n=1, prompt='The coronavirus is', max_length=100, seed=12)
print('-----')
ai.generate(n=1, prompt='What is Black Lives Matter?', max_length=100, seed=13)
print('-----')
ai.generate(n=1, prompt='What is white supremacy?', max_length=100, seed=14)
````

GTP-2 ble av OpenAI lansert i februar 2019 (se [1]), legg derfor spesielt merke til hvordan den både genererer for "What is Black Lives Matter?" og "The coronavirus is", da begge disse hendelsene først kom i 2020. For koronaviruset virker det som den bare utbroderer fra generelt et virus, og for _Black Lives Matter_ virker det som den blir, av en litt merkelig grunn, veldig rasistisk og påstår nærmest det motsatte av hva _Black Lives Matter_ står for. Det skal riktignok bemerkes at dette bare er et generert tilfelle, og at det ikke er gitt at den produserer like "drøye" påstander med en annen _seed_. 

Det språkmodellen kanskje klarer best, er klimaendringer, hvor den (som vi også skal se i oppgave b) henvender seg til reelle (om ikke helt eksakte) endringer i naturen. Naturlig nok er den også trent på mye stoff om daværende amerikanske president Donald Trump, noe som synes i den mer eller mindre gode kvaliteten i den genererte teksten.
